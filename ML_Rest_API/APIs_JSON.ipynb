### **What is JSON?**

JSON, or JavaScript Object Notation, is a lightweight and human-readable format used to represent and exchange data. It is text-based and easy to work with, making it a universal choice for data transmission, especially in web applications. Despite its roots in JavaScript, JSON is language-agnostic, meaning it can be utilized in almost any programming environment.


### **Why JSON?**

The strength of JSON lies in its simplicity. Its syntax is straightforward, yet it is powerful enough to represent complex data structures. JSON is compact, which makes it ideal for transferring data over networks. Furthermore, its human-readable format allows developers to debug and understand data without requiring specialized tools.

JSON's popularity stems from its versatility—it can be used for everything from configuring systems to exchanging data with APIs and storing information in NoSQL databases.





### **The Structure of JSON**

#### **Objects: Key-Value Pairs**
At its core, JSON data is built using **objects**. An object is a collection of key-value pairs, where the keys are always strings enclosed in double quotes, and the values can be strings, numbers, booleans, null, other objects, or arrays. Objects are enclosed in curly braces (`{ }`).

Here’s a simple example:

```json
{
  "name": "Alice",
  "age": 25,
  "is_student": true
}
```

In this example:
- `"name"` is the key, paired with the string value `"Alice"`.
- `"age"` is a key paired with a numeric value, `25`.
- `"is_student"` is a key paired with a boolean value, `true`.

#### **Arrays: Ordered Lists**
JSON also supports **arrays**, which are ordered lists of values enclosed in square brackets (`[ ]`). Arrays can hold multiple values, which can be strings, numbers, booleans, objects, or even other arrays.

Here’s an example of an array in JSON:

```json
{
  "subjects": ["Math", "Science", "History"]
}
```

The `subjects` key maps to an array containing three string values.

#### **Nested Data**
JSON allows for nesting, enabling you to represent complex, hierarchical data structures. For example:

```json
{
  "name": "Alice",
  "age": 25,
  "address": {
    "street": "123 Maple St",
    "city": "Wonderland"
  },
  "subjects": ["Math", "Science", "History"]
}
```

Here, the `address` key maps to another JSON object, and `subjects` maps to an array. This nested structure makes JSON flexible enough to represent real-world data like user profiles, product catalogs, or API responses.


### **JSON Syntax Rules**

To maintain its simplicity, JSON follows strict syntax rules:
1. **Key-Value Format**: Data is organized into key-value pairs, separated by colons (`:`).
2. **Keys Are Strings**: Keys must always be enclosed in double quotes (`" "`).
3. **Valid Data Types**: Values can be:
   - Strings (e.g., `"Alice"`)
   - Numbers (e.g., `25`)
   - Booleans (`true` or `false`)
   - Null (`null`)
   - Objects
   - Arrays
4. **No Trailing Commas**: Commas are used to separate pairs, but trailing commas are not allowed.
5. **Case Sensitivity**: Both keys and values in JSON are case-sensitive.












### **1. Converting JSON Strings to Python Dictionaries**
The `json` module in Python is used to work with JSON. The `json.loads` method converts a JSON string into a Python dictionary.













````python
import json

# JSON string
json_string = '{"name": "Alice", "age": 25, "is_student": true}'

# Convert JSON string to dictionary
data = json.loads(json_string)

# Access values
print(data["name"])
print(data["age"])
print(data["is_student"])
````

### **2. Converting Python Dictionaries to JSON Strings**
The `json.dumps` method converts a Python dictionary to a JSON-formatted string.


````python
import json

# Python dictionary
python_dict = {
    "name": "Bob",
    "age": 30,
    "skills": ["Python", "JavaScript"],
    "is_employed": False
}

# Convert dictionary to JSON string
json_string = json.dumps(python_dict, indent=4)

print(json_string)
````

### **3. Reading JSON from a File**
You can load JSON data from a file using the `json.load` method.

Suppose you have a file named `data.json` with the following content:
```json
{
    "name": "Alice",
    "age": 25,
    "skills": ["Math", "Science"],
    "is_student": true
}


````python
import json

# Open and read the JSON file
with open('data.json', 'r') as file:
    data = json.load(file)

# Access values
print(data["name"])
print(data["skills"])
````

### **4. Writing JSON to a File**
You can save a Python dictionary as a JSON file using `json.dump`.



````python
import json

# Python dictionary
data = {
    "name": "Charlie",
    "age": 28,
    "skills": ["HTML", "CSS", "JavaScript"],
    "is_employed": True
}

# Write to a JSON file
with open('output.json', 'w') as file:
    json.dump(data, file, indent=4)

print("Data written to output.json")
````

### **5. Nested JSON and Dictionary Manipulation**
You can work with nested JSON data like you would with nested dictionaries in Python.

````python
import json

# JSON string with nested objects
json_string = '''
{
    "name": "Alice",
    "address": {
        "street": "123 Maple St",
        "city": "Wonderland",
        "zip": "12345"
    },
    "skills": ["Python", "Data Analysis"]
}
'''

# Convert to Python dictionary
data = json.loads(json_string)

# Access nested values
print(data["address"]["city"])
print(data["skills"][1])

# Modify a value
data["address"]["zip"] = "67890"

# Convert back to JSON
updated_json = json.dumps(data, indent=4)
print(updated_json)
````


### **6. Merging and Updating JSON-like Dictionaries**
You can merge or update JSON data directly, as they are represented as Python dictionaries.


````python
# Two dictionaries
data1 = {"name": "Alice", "age": 25}
data2 = {"skills": ["Python", "Machine Learning"], "is_student": True}

# Merge dictionaries
merged_data = {**data1, **data2}
print(merged_data)

# Update a dictionary
data1.update(data2)
print(data1)
````

### **8. Handling Errors in JSON Operations**
You should handle errors when working with JSON data to prevent crashes.



````python
import json

# Invalid JSON string
invalid_json = '{"name": "Alice", "age": 25,}'  # Trailing comma causes error

try:
    data = json.loads(invalid_json)
except json.JSONDecodeError as e:
    print(f"Failed to parse JSON: {e}")
````

### **Background on the OpenWeather API**

The OpenWeather API is a widely used web service that provides global weather data to developers and organizations. It is a product of **OpenWeather**, a company that specializes in collecting, processing, and delivering weather-related data. This API is particularly popular for building weather-based applications, dashboards, and analysis tools.

#### **Data Sources Used by OpenWeather**

OpenWeather aggregates its data from multiple sources, including:
1. **Weather Stations**:
   - Data collected from over 40,000 weather stations worldwide.
2. **Satellite Imagery**:
   - Provides atmospheric observations and cloud coverage data.
3. **Meteorological Models**:
   - Combines historical and real-time data with forecasting models to predict weather trends.
4. **IoT Sensor Networks**:
   - Contributions from private networks of sensors and devices.

#### **Types of APIs Provided**

The OpenWeather API offers several specialized APIs to cater to different use cases:

1. **Current Weather Data API**
   - Provides current weather conditions for a city or geographic coordinates.
   - Key parameters: temperature, wind speed, humidity, weather description, and more.

2. **5-Day/3-Hour Forecast API**
   - Delivers weather forecasts for the next 5 days, with data points every 3 hours.

3. **One Call API**
   - Combines current weather, daily/hourly forecasts, and historical data into a single request.
   - Includes optional data like UV index, moon phase, and air quality.

4. **Historical Weather API**
   - Offers historical weather data for a given location. Useful for research and trend analysis.

5. **Air Pollution API**
   - Provides air quality index (AQI) and pollutant concentrations for specific locations.

#### **1. Current Weather Data Endpoint**
Fetch real-time weather data for a city.

````python
#https://api.openweathermap.org/data/3.0/onecall?lat={lat}&lon={lon}&exclude={part}&appid={API key}

import requests

# OpenWeather API details
BASE_URL = "https://api.openweathermap.org/data/2.5/weather"

# Function to fetch weather data
def get_weather(city):
    # Define parameters
    params = {
        "q": city,
        "appid": API_KEY,
        "units": "metric"  # Use "imperial" for Fahrenheit
    }

    # Send GET request
    response = requests.get(BASE_URL, params=params)

    # Check if the response is valid
    if response.status_code == 200:
        data = response.json()  # Parse JSON response
        return data
    else:
        print(f"Error: Unable to fetch weather data. HTTP {response.status_code}")
        return None

# Fetch weather for a specific city
city = "London"
weather_data = get_weather(city)

# Display the raw JSON response (optional for debugging)
if weather_data:
    print("Raw JSON Response:")
    print(weather_data)
````

#### **2. 5-Day/3-Hour Forecast Endpoint**
Fetch a 5-day weather forecast with 3-hour intervals.

````python
BASE_URL = "https://api.openweathermap.org/data/2.5/forecast"

def get_forecast(city):
    params = {
        "q": city,
        "appid": API_KEY,
        "units": "imperial"
    }
    response = requests.get(BASE_URL, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}")
        return None

city = "Tokyo"
forecast_data = get_forecast(city)
if forecast_data:
    print(f"5 Step, 3 hours Forecast for {city}:")
    for item in forecast_data["list"][:5]:  # Print first 5 intervals
        print(f"Date: {item['dt_txt']}, Temperature: {item['main']['temp']}°F, Weather: {item['weather'][0]['main']} - {item['weather'][0]['description']}")
````

#### **3. Air Pollution API Endpoint**
Fetch air quality data for a specific location.

````python

BASE_URL = "https://api.openweathermap.org/data/2.5/air_pollution"

def get_air_quality(lat, lon):
    params = {
        "lat": lat,
        "lon": lon,
        "appid": API_KEY
    }
    response = requests.get(BASE_URL, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}")
        return None

# Example: Air quality for Bengaluru, India
lat, lon = 12.9716, 77.5946
air_quality_data = get_air_quality(lat, lon)
if air_quality_data:
    aqi = air_quality_data["list"][0]["main"]["aqi"]
    print(f"Air Quality Index (AQI): {aqi}")
````

While JSON is widely adopted, it does have some limitations:
1. **No Comments**: Unlike some data formats, JSON does not support comments, which can make documentation within the data more challenging.
2. **Limited Data Types**: JSON lacks support for more complex data types like dates, which are often represented as strings.
3. **Verbosity**: Though compact, JSON can still become verbose compared to binary formats like Protocol Buffers.


### **2. Nested JSON Parsing**
**Context**: Work with nested JSON data, such as a detailed user profile or nested API response.

**Challenge**:
Given the JSON object below representing a user profile:
1. Print the user's full name (combine first and last name).
2. Print the name of the user's first favorite book.
3. List all skills along with their levels.

**Sample JSON Data**:
```json
{
    "user": {
        "first_name": "Jane",
        "last_name": "Doe",
        "profile": {
            "age": 30,
            "email": "jane.doe@example.com",
            "favorites": {
                "books": [
                    "Pride and Prejudice",
                    "To Kill a Mockingbird"
                ]
            }
        },
        "skills": {
            "Python": "Advanced",
            "JavaScript": "Intermediate",
            "SQL": "Beginner"
        }
    }
}
```

**Expected Output**:
```plaintext
Full Name: Jane Doe
First Favorite Book: Pride and Prejudice
Skills:
- Python: Advanced
- JavaScript: Intermediate
- SQL: Beginner
```


````python
import json

# Sample JSON
user_data = '''
{
    "user": {
        "first_name": "Jane",
        "last_name": "Doe",
        "profile": {
            "age": 30,
            "email": "jane.doe@example.com",
            "favorites": {
                "books": [
                    "Pride and Prejudice",
                    "To Kill a Mockingbird"
                ]
            }
        },
        "skills": {
            "Python": "Advanced",
            "JavaScript": "Intermediate",
            "SQL": "Beginner"
        }
    }
}
'''

def user_info(data):
  data = json.loads(data)
  print(f"Full Name: {data['user']['first_name']} {data['user']['last_name']}")
  print(f"First Favorite Book: {data['user']['profile']['favorites']['books'][0]}")
  print(f"Skills:")
  for skill, level in data['user']['skills'].items():
    print(f"- {skill}: {level}")

user_info(user_data)
````
